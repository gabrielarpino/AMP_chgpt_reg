# Posterior over change points in the Linear Model

```python
import numpy as np
import jax.numpy as jnp # Not used in any randomness, all randomness is in numpy
import jax
import numpy.random as nprandom
import matplotlib.pyplot as plt
from amp import MSE, norm_sq_corr, PAL, hausdorff
from amp import comparison
from amp.marginal_separable_jax import q, GAMP_full, SE_fixed_C_v1, GAMP_gaussian_clean, SE_fixed_C, GAMP
import amp.marginal_separable_jax
import amp.changepoint_jax as changepoint_jax
import amp.signal_configuration
from amp.signal_configuration import C_to_chgpt, generate_C_distanced, unif_prior_to_η_ϕ
from amp.signal_priors import SignalPrior, SparseDiffSignal, SparseGaussianSignal, GaussianSignal
from amp.posterior import MAP, MAP_η, compute_posterior, posterior_over_η, η_to_ψ
import amp.posterior
from amp.marginal_separable_jax import GAMP_gaussian_clean, GAMP_real_data, q
from amp.signal_priors import GaussianSignal
import amp.gof
from functools import partial
from tqdm import tqdm
import scipy
tqdm = partial(tqdm, position=0, leave=True)

np.random.seed(6)

p = 800
σ = 0.1 # noise standard deviation
L = 3 # num signals. L-1 changepoints

T = 10 # num_iterations
δ_list = [0.5, 0.75, 1.0]
# δ_list = [1.0, 1.5]
num_trials = 30

fractional_chgpt_locations = np.array([1/3, 8/15])
# fractional_chgpt_locations = 0.4

corr_1_list = [[] for _ in range(len(δ_list))]
corr_2_list = [[] for _ in range(len(δ_list))]
corr_1_SE_list = [[] for _ in range(len(δ_list))]
corr_2_SE_list = [[] for _ in range(len(δ_list))]
fractional_η̂_list = [[] for _ in range(len(δ_list))]
amp_post_list = [[] for _ in range(len(δ_list))]
theory_post_list = [[] for _ in range(len(δ_list))]

for i in tqdm(range(len(δ_list))):
    δ = δ_list[i]
    print("δ = ", δ)

    n = int(δ * p)
    true_chgpt_locations = lambda n: (n*fractional_chgpt_locations).astype(int)

    # --- Preliminaries ---
    # Changepoint locations
    Δ = lambda n: int(n/5)
    Lmin = 3
    Lmax = 3
    subsample_factor = 10
    n_subsampled = int(n/subsample_factor)
    num_valid_configs, η_arr, p_η_arr, ϕ = unif_prior_to_η_ϕ(Lmin, Lmax, Δ(n), n) # Lxn matrix
    _, η_arr_subsampled, p_η_arr_subsampled, _ = unif_prior_to_η_ϕ(Lmin, Lmax, Δ(n_subsampled), n_subsampled) # Lxn matrix
    η_arr_subsampled = η_arr_subsampled * subsample_factor # Rescale η_arr_subsampled to match η_arr
    assert np.all([-1 not in η_arr[i] for i in range(η_arr.shape[0])]) # Assert that there are no -1 values in η_arr, as \eta_to_ψ_jax assumes this
    assert η_arr.shape == (num_valid_configs, Lmax-1) and \
        len(p_η_arr) == num_valid_configs
    assert ϕ.shape == (Lmax, n)

    print(f"True chgpts = {true_chgpt_locations(n)}")
    η_true = true_chgpt_locations(n)
    C_true = η_to_ψ(η_true, n)

    ν_fixed_tabulated, κ_T_fixed_tabulated = None, None

    for trial_num in tqdm(range(num_trials)):

        # Gaussian Signal Prior
        B̃_cov = np.eye(L)
        B̃_cov[0, 1] = 0.0
        B̃_cov[1, 0] = 0.0
        signal_prior = GaussianSignal(B̃_cov)

        # Sparse Gaussian Signal Prior
        # α = 0.6 # fraction of nonzeros
        # σ_l_arr = np.array([0.5, 0.5]) # signal standard deviation
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)

        # Sparse Difference Signal Prior
        # α = 0.25 # fraction of nonzeros
        # # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # var_β_1 = 1
        # σ_w = 1
        # signal_prior = SparseDiffSignal(var_β_1, σ_w, α, L)

        # Soft threshold denoiser 
        # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # α = 0.5 # fraction of nonzeros
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)
        # st_ζ = 0.5 # The scaling factor in front of the effective noise std in the threshold
        # if soft thresholding. Setting st_ζ != None in the GAMP_full function.

        B̃ = signal_prior.sample(p)
        B̂_0 = signal_prior.sample(p)

        ρ = 1/δ * 1/p * B̃.T @ B̃

        # Generate the observation vector Y
        X = np.random.normal(0, np.sqrt(1/n), (n, p))
        Θ = X @ B̃
        Y = q(Θ, C_true, σ) 

        B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr, ν_fixed_arr, κ_T_fixed_arr, ν̂_fixed_arr, κ_B_fixed_arr = GAMP_full(B̃, δ, p, ϕ, σ, X, Y, C_true, T, B̂_0, signal_prior, signal_prior, st_ζ = None, verbose = False, seed = None, tqdm_disable = False)
        # B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr = GAMP(B̃, δ, p, ϕ, σ, X, Y, T, true_signal_prior=signal_prior, est_signal_prior=signal_prior)
        ν = ν_arr[-1]
        κ_T = κ_T_arr[-1]
        ν_fixed = ν_fixed_arr[-1]
        κ_T_fixed = κ_T_fixed_arr[-1]

        # (δ * ρ[1, 1] - δ * ν_arr[-1, 1, 1]) + (δ * ρ[0, 0] - δ * ν_arr[-1, 0, 0]) - 2*(δ * ρ[0, 1] - δ * ν_arr[-1, 0, 1])
        # ν_fixed, κ_T_fixed = SE_fixed_C(C_true, signal_prior, signal_prior, δ, p, ϕ, L, σ, T, tqdm_disable = False, st_ζ=None)

        corr_1_list[i].append(norm_sq_corr(B̂[:, 0], B̃[:, 0]))
        corr_2_list[i].append(norm_sq_corr(B̂[:, 1], B̃[:, 1]))
        corr_1_SE_list[i].append(ν_fixed[0, 0]**2 / ( ρ[0, 0] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[0, 0]))
        corr_2_SE_list[i].append(ν_fixed[1, 1]**2 / ( ρ[1, 1] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[1, 1]))

        # Subsample η_arr, p_η_arr
        # subsample_factor = 200
        # η_arr_sub = η_arr[0::subsample_factor]
        # p_η_arr_sub = p_η_arr[0::subsample_factor]


        ### --- Compute the posterior from AMP --- ###
        print("Computing AMP posterior...")
        if ν_fixed_tabulated is None or κ_T_fixed_tabulated is None: # Only tabulate these values once. 
            ν_fixed_tabulated, κ_T_fixed_tabulated = amp.posterior.tabulate_state_evolution(η_arr_subsampled, signal_prior, signal_prior, \
                                        δ, p, ϕ, L, σ, T, \
                                        ν_arr, κ_T_arr, ν̂_arr, κ_B_arr, st_ζ = None, tqdm_disable = False)
        amp_post = amp.posterior.exact_posterior(η_arr_subsampled, p_η_arr_subsampled, Θ_t, Y, ρ, δ, ν_fixed_tabulated, κ_T_fixed_tabulated)
        amp_post_list[i].append(amp_post)
        print("Posterior computed.")

        ### --- Compute the posterior from theory --- ###
        print("Computing theory posterior...")
        Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), ρ, size=n))).reshape(n, L)
        # Sample V̄_Θ
        V_Θ = (Z @ np.linalg.inv(ρ) @ ν_fixed + jnp.array(nprandom.multivariate_normal(
            jnp.zeros(L), κ_T_fixed, size=n))).reshape(n, L)
        # # post_V_q = posterior_over_η(η_arr, p_η_arr, V_Θ_i, q(Z_i, C_true, σ), jnp.array(ρ), σ, ν, κ_T)
        theory_post = amp.posterior.exact_posterior(η_arr_subsampled, p_η_arr_subsampled, V_Θ, q(Z, C_true, σ), ρ, δ, ν_fixed_tabulated, κ_T_fixed_tabulated)
        theory_post_list[i].append(theory_post)
        print("Theory posterior computed.")

        # Convert to marginal
        C_s_sub = jax.vmap(amp.posterior.η_to_ψ_jax, in_axes = (0, None))(η_arr_subsampled, n)
        amp_marg_prob = amp.signal_configuration.marginal_prob_chgpt(C_s_sub, amp_post)
        theory_marg_prob = amp.signal_configuration.marginal_prob_chgpt(C_s_sub, theory_post)

        plt.plot(amp_marg_prob, label="AMP")
        plt.plot(theory_marg_prob, label="theory")
        plt.title("Posterior")
        plt.legend()
        plt.show()


        # # --- Compute η̂ from theoretical ---
        # Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), ρ, size=n))).reshape(n, L)
        # # Sample V̄_Θ
        # V_Θ = (Z @ np.linalg.inv(ρ) @ ν_fixed + jnp.array(nprandom.multivariate_normal(
        #     jnp.zeros(L), κ_T_fixed, size=n))).reshape(n, L)
        # # post_V_q = posterior_over_η(η_arr, p_η_arr, V_Θ_i, q(Z_i, C_true, σ), jnp.array(ρ), σ, ν, κ_T)

        # η̂_theory, gof_list_theory = amp.gof.argmax_gof(η_arr, V_Θ @ np.linalg.inv(np.linalg.inv(ρ) @ ν_fixed), q(Z, C_true, σ), model = 'linear')
        # theory_hausdorff[i].append((1/n) * amp.performance_measures.hausdorff(η̂_theory, η_true))

# Fix for tikzplotlib
from matplotlib.lines import Line2D
from matplotlib.legend import Legend
Line2D._us_dashSeq    = property(lambda self: self._dash_pattern[1])
Line2D._us_dashOffset = property(lambda self: self._dash_pattern[0])
Legend._ncol = property(lambda self: self._ncols)

# Sqcorr plots
plt.errorbar(δ_list, np.mean(np.array(corr_1_list), axis = 1), yerr = np.std(np.array(corr_1_list), axis = 1) , label="AMP β¹", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np.array(corr_2_list), axis = 1),yerr = np.std(np.array(corr_2_list), axis = 1) ,  label="AMP β²", marker="o", ls="--", c=PAL[2], alpha = 0.75)
plt.errorbar(δ_list,np.mean(np.array(corr_1_SE_list), axis = 1),yerr = np.std(np.array(corr_1_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[0])
plt.errorbar(δ_list, np.mean(np.array(corr_2_SE_list), axis = 1),yerr = np.std(np.array(corr_2_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[2])
plt.ylim((0, 1))
plt.xlabel("δ")
plt.ylabel("Norm Sq Corr")
plt.legend()
plt.show()

# plt.savefig("figures/synthetic_estimation_linear_3chgpt_" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + ".pdf")
# import tikzplotlib
# tikzplotlib.save("figures/synthetic_estimation_linear_3chgpt_" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + ".pdf")
```
