```python
import numpy as np
import jax.numpy as jnp # Not used in any randomness, all randomness is in numpy
import jax
import numpy.random as nprandom
import matplotlib.pyplot as plt
from amp import MSE, norm_sq_corr, PAL, hausdorff
from amp import comparison
from amp.marginal_separable_jax import q, GAMP_full, SE_fixed_C_v1, GAMP_gaussian_clean, SE_fixed_C, GAMP
import amp.marginal_separable_jax
import amp.changepoint_jax as changepoint_jax
import amp.signal_configuration
from amp.signal_configuration import C_to_chgpt, generate_C_distanced, unif_prior_to_Î·_Ï, unif_prior_to_Î·_Ï_combinations
from amp.signal_priors import SignalPrior, SparseDiffSignal, SparseGaussianSignal, GaussianSignal
from amp.posterior import MAP, MAP_Î·, posterior_over_Î·
from amp.signal_configuration import Î·_to_Ï
from amp.marginal_separable_jax_linear import GAMP_gaussian_clean, GAMP_real_data, q
from amp.signal_priors import GaussianSignal
import amp.posterior
import amp.gof
from functools import partial
from tqdm import tqdm
import scipy
tqdm = partial(tqdm, position=0, leave=True)

np.random.seed(5)

p = 500 
Ï = 0.2 # noise standard deviation
L = 4 # num signals. L-1 changepoints

T = 10 # num_iterations
Î´_list = [0.5, 1.0, 1.5]
# Î´_list = [0.75, 1.5]
num_trials = 5

# fractional_chgpt_locations = np.array([0.3, 0.7])
fractional_chgpt_locations = np.array([0.6])
# fractional_chgpt_locations = np.array([])
# fractional_chgpt_locations = 0.4

corr_1_list = [[] for _ in range(len(Î´_list))]
corr_2_list = [[] for _ in range(len(Î´_list))]
corr_1_SE_list = [[] for _ in range(len(Î´_list))]
corr_2_SE_list = [[] for _ in range(len(Î´_list))]
fractional_Î·Ì_list = [[] for _ in range(len(Î´_list))]
amp_hausdorff = [[] for _ in range(len(Î´_list))]
theory_hausdorff = [[] for _ in range(len(Î´_list))]
amp_size = [[] for _ in range(len(Î´_list))]
theory_size = [[] for _ in range(len(Î´_list))]

for i in tqdm(range(len(Î´_list))):
    Î´ = Î´_list[i]
    print("Î´ = ", Î´)

    n = int(Î´ * p)
    true_chgpt_locations = lambda n: (n*fractional_chgpt_locations).astype(int)

    # --- Preliminaries ---
    # Changepoint locations
    Î = lambda n: int(n/5)
    Lmax = L
    Lmin = 1
    p_l = 1/L * np.ones(L) # puts equal weight on all number of signals, set None otherwise. Best for estimating size. 
    num_valid_configs, Î·_arr, p_Î·_arr, Ï = unif_prior_to_Î·_Ï_combinations(Lmin, Lmax, Î(n), n, p_l = p_l) # Lxn matrix
    # Ï = Ï / Ï.sum(axis = 1)[:, None]
    plt.plot(Ï.T)
    plt.title("Marginal")
    plt.show()
    print(" sum:", Ï.sum(axis = 1))
    # Lmin = L
    # p_l = None
    # num_valid_configs, Î·_arr, p_Î·_arr, Ï = unif_prior_to_Î·_Ï(Lmin, Lmax, Î(n), n, p_l = p_l) # Lxn matrix

    # Optional subsampling to speed up computation
    subsample_factor = 1 # Increasing this to 10 makes estimation performance much worse, but is good for posterior plotting. 
    n_subsampled = int(n/subsample_factor)
    _, Î·_arr_subsampled, p_Î·_arr_subsampled, _ = unif_prior_to_Î·_Ï_combinations(Lmin, Lmax, Î(n_subsampled), n_subsampled, p_l = p_l) # Lxn matrix
    # _, Î·_arr_subsampled, p_Î·_arr_subsampled, _ = unif_prior_to_Î·_Ï(Lmin, Lmax, Î(n_subsampled), n_subsampled, p_l = p_l) # Lxn matrix
    Î·_arr_subsampled = Î·_arr_subsampled * subsample_factor # Rescale Î·_arr_subsampled to match Î·_arr

    assert Î·_arr.shape == (num_valid_configs, Lmax-1) and \
        len(p_Î·_arr) == num_valid_configs
    assert Ï.shape == (Lmax, n)

    print(f"True chgpts = {true_chgpt_locations(n)}")
    Î·_true = true_chgpt_locations(n)
    if len(Î·_true) > 0:
        C_true = Î·_to_Ï(Î·_true, n)
    else:
        C_true = np.zeros((n, ))

    for trial_num in tqdm(range(num_trials)):

        # Gaussian Signal Prior
        BÌ_cov = 1 * np.eye(L)
        BÌ_cov[0, 1] = 0.2
        BÌ_cov[1, 0] = 0.2
        signal_prior = GaussianSignal(BÌ_cov)

        # Sparse Gaussian Signal Prior
        # Î± = 0.6 # fraction of nonzeros
        # Ï_l_arr = np.array([0.5, 0.5]) # signal standard deviation
        # signal_prior = SparseGaussianSignal(Î±, Î´, Ï_l_arr)

        # Sparse Difference Signal Prior
        # Î± = 0.25 # fraction of nonzeros
        # # Ï_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # var_Î²_1 = 1
        # Ï_w = 1
        # signal_prior = SparseDiffSignal(var_Î²_1, Ï_w, Î±, L)

        # Soft threshold denoiser 
        # Ï_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # Î± = 0.5 # fraction of nonzeros
        # signal_prior = SparseGaussianSignal(Î±, Î´, Ï_l_arr)
        # st_Î¶ = 0.5 # The scaling factor in front of the effective noise std in the threshold
        # if soft thresholding. Setting st_Î¶ != None in the GAMP_full function.

        BÌ = signal_prior.sample(p)
        BÌ_0 = signal_prior.sample(p)

        Ï = 1/Î´ * 1/p * BÌ.T @ BÌ

        # Generate the observation vector Y
        X = np.random.normal(0, np.sqrt(1/n), (n, p))
        Î = X @ BÌ
        Îµ = np.random.randn(n) * Ï # Random normal!!! not uniform
        Y = q(Î, C_true, Îµ) 

        BÌ, Î_t, Î½_arr, Îº_T_arr, Î½Ì_arr, Îº_B_arr, Î½_fixed_arr, Îº_T_fixed_arr, Î½Ì_fixed_arr, Îº_B_fixed_arr = GAMP_full(BÌ, Î´, p, Ï, Ï, X, Y, C_true, T, BÌ_0, signal_prior, signal_prior, st_Î¶ = None, verbose = False, seed = None, tqdm_disable = False)
        # BÌ, Î_t, Î½_arr, Îº_T_arr, Î½Ì_arr, Îº_B_arr = GAMP(BÌ, Î´, p, Ï, Ï, X, Y, T, true_signal_prior=signal_prior, est_signal_prior=signal_prior)
        Î½ = Î½_arr[-1]
        Îº_T = Îº_T_arr[-1]
        Î½_fixed = Î½_fixed_arr[-1]
        Îº_T_fixed = Îº_T_fixed_arr[-1]

        # (Î´ * Ï[1, 1] - Î´ * Î½_arr[-1, 1, 1]) + (Î´ * Ï[0, 0] - Î´ * Î½_arr[-1, 0, 0]) - 2*(Î´ * Ï[0, 1] - Î´ * Î½_arr[-1, 0, 1])
        # Î½_fixed, Îº_T_fixed = SE_fixed_C(C_true, signal_prior, signal_prior, Î´, p, Ï, L, Ï, T, tqdm_disable = False, st_Î¶=None)

        corr_1_list[i].append(norm_sq_corr(BÌ[:, 0], BÌ[:, 0]))
        corr_2_list[i].append(norm_sq_corr(BÌ[:, 1], BÌ[:, 1]))
        corr_1_SE_list[i].append(Î½_fixed[0, 0]**2 / ( Ï[0, 0] * (Îº_T_fixed + Î½_fixed.T @ np.linalg.inv(Ï) @ Î½_fixed)[0, 0]))
        corr_2_SE_list[i].append(Î½_fixed[1, 1]**2 / ( Ï[1, 1] * (Îº_T_fixed + Î½_fixed.T @ np.linalg.inv(Ï) @ Î½_fixed)[1, 1]))

        ### --- Approx posterior --- ###
        # Compute approximate posterior MAP
        print("Beginning approximate posterior computation...")
        Ï_arr_amp, Î·_idx_amp = amp.signal_configuration.Î·_to_Ï_jax_combinations_mapped(Î·_arr_subsampled, n, L)       
        post_amp = amp.posterior.compute_approx_posterior(Ï_arr_amp, Î_t, Y, n, Ï, Ï, Î½, Îº_T)
        map_idx_amp = np.argmax(post_amp)
        ÏÌ = Ï_arr_amp[map_idx_amp]
        Î·Ì = Î·_arr_subsampled[Î·_idx_amp[map_idx_amp]]
        Î·Ì = Î·Ì[Î·Ì > 0]

        fractional_Î·Ì_list[i].append(Î·Ì / n)
        amp_size[i].append(len(Î·Ì))
        amp_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(Î·Ì, Î·_true), 0.5)) # Sometimes infs occur because no change point is returned

        # --- Compute Î·Ì from theoretical ---
        Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), Ï, size=n))).reshape(n, L)
        # Sample VÌ_Î
        V_Î = (Z @ np.linalg.inv(Ï) @ Î½_fixed + jnp.array(nprandom.multivariate_normal(
            jnp.zeros(L), Îº_T_fixed, size=n))).reshape(n, L)
        # post_V_q = posterior_over_Î·(Î·_arr, p_Î·_arr, V_Î_i, q(Z_i, C_true, Ï), jnp.array(Ï), Ï, Î½, Îº_T)

        Ï_arr_theory, Î·_idx_theory = amp.signal_configuration.Î·_to_Ï_jax_combinations_mapped(Î·_arr_subsampled, n, L)       
        post_theory = amp.posterior.compute_approx_posterior(Ï_arr_theory, V_Î, q(Z, C_true, Îµ), n, Ï, Ï, Î½, Îº_T)
        map_idx_theory = np.argmax(post_theory)
        ÏÌ_theory = Ï_arr_theory[map_idx_theory]
        Î·Ì_theory = Î·_arr_subsampled[Î·_idx_theory[map_idx_theory]]
        Î·Ì_theory = Î·Ì_theory[Î·Ì_theory > 0] # Remove negative ones, which are invalid

        theory_size[i].append(len(Î·Ì_theory))
        theory_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(Î·Ì_theory, Î·_true), 0.5)) # Sometimes infs occur because no change point is returned
        print("Approximate posterior computation finished.")

        ### --- GOF No Combinations --- ###
        # Î·Ì, gof_list = amp.gof.argmax_gof(Î·_arr, Î_t, Y, model = 'linear')
        # plt.plot(gof_list, label="AMP")
        # plt.title("GOF AMP")
        # plt.legend()
        # plt.show()
        # print("Î·Ì = ", Î·Ì)

        # fractional_Î·Ì_list[i].append(Î·Ì / n)
        # amp_hausdorff[i].append((1/n) * amp.performance_measures.hausdorff(Î·Ì, Î·_true))

        # # --- Compute Î·Ì from theoretical ---
        # Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), Ï, size=n))).reshape(n, L)
        # # Sample VÌ_Î
        # V_Î = (Z @ np.linalg.inv(Ï) @ Î½_fixed + jnp.array(nprandom.multivariate_normal(
        #     jnp.zeros(L), Îº_T_fixed, size=n))).reshape(n, L)
        # # post_V_q = posterior_over_Î·(Î·_arr, p_Î·_arr, V_Î_i, q(Z_i, C_true, Ï), jnp.array(Ï), Ï, Î½, Îº_T)

        # Î·Ì_theory, gof_list_theory = amp.gof.argmax_gof(Î·_arr, V_Î, q(Z, C_true, Ï), model = 'linear')
        # plt.plot(gof_list_theory, label="Theory", ls="--")
        # plt.title("GOF Theory")
        # plt.legend()
        # plt.show()
        # print("Î·Ì_theory = ", Î·Ì_theory)
        # theory_hausdorff[i].append((1/n) * amp.performance_measures.hausdorff(Î·Ì_theory, Î·_true))

        fig = plt.figure()
        plt.xlim((0, n))
        plt.plot(np.arange(n), ÏÌ, label="AMP")
        plt.plot(np.arange(n), ÏÌ_theory, label="Theory", ls="--")
        [plt.axvline(x=loc, color='r', linestyle='-.', alpha = 0.5, label="AMP chgpt") for loc in Î·Ì]
        [plt.axvline(x=loc, color='y', linestyle='-.', alpha = 0.5, label="Theory chgpt") for loc in Î·Ì_theory]
        [plt.axvline(x=c, color='black', linestyle='--', alpha = 0.3, label="True") for c in true_chgpt_locations(n)]

        # plt.plot(gof_list, label="AMP")
        # plt.plot(gof_list_theory, label="Theory", ls="--")
        # for chgpt_loc in true_chgpt_locations(n):
        #     plt.axvline(chgpt_loc - Î(n), color='r', linestyle='-', label="True changepoint location")
        # plt.title("GOF")
        plt.legend()
        plt.show()

# Fix for tikzplotlib
from matplotlib.lines import Line2D
from matplotlib.legend import Legend
Line2D._us_dashSeq    = property(lambda self: self._dash_pattern[1])
Line2D._us_dashOffset = property(lambda self: self._dash_pattern[0])
Legend._ncol = property(lambda self: self._ncols)

# Sqcorr plots
plt.errorbar(Î´_list, np.mean(np.array(corr_1_list), axis = 1), yerr = np.std(np.array(corr_1_list), axis = 1) , label="AMP Î²Â¹", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(Î´_list, np.mean(np.array(corr_2_list), axis = 1),yerr = np.std(np.array(corr_2_list), axis = 1) ,  label="AMP Î²Â²", marker="o", ls="--", c=PAL[2], alpha = 0.75)
plt.errorbar(Î´_list,np.mean(np.array(corr_1_SE_list), axis = 1),yerr = np.std(np.array(corr_1_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[0])
plt.errorbar(Î´_list, np.mean(np.array(corr_2_SE_list), axis = 1),yerr = np.std(np.array(corr_2_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[2])
plt.ylim((0, 1))
plt.xlabel("Î´")
plt.ylabel("Norm Sq Corr")
plt.legend()
plt.show()

# Î·Ì plots
# plt.plot(Î´_list, fractional_Î·Ì_list, label="fractional Î·Ì")
# # Plot the true changepoint location, horizontal line
# plt.axhline(y=fractional_chgpt_locations, color='r', linestyle='-', label="True changepoint location")
# plt.ylim((0, 1))
# plt.xlabel("Î´")
# plt.ylabel("Î·Ì")
# plt.legend()
# plt.show()

# Hausdorff plots
np_amp_hausdorff = np.array(amp_hausdorff)
np_theory_hausdorff = np.array(theory_hausdorff)
plt.errorbar(Î´_list, np.mean(np_amp_hausdorff, axis = 1), yerr = np.mean(np.array(amp_hausdorff), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(Î´_list, np.mean(np_theory_hausdorff, axis = 1),yerr = np.mean(np.array(theory_hausdorff), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 0.5))
plt.xlabel("Î´")
plt.ylabel("Hausdorff distance")
plt.legend()
plt.savefig("figures/synthetic_estimation_linear_L4_" + "Ï_" + str(Ï) + "_num_trials_" + str(num_trials) + "theta_T_3.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_estimation_linear_L4_" + "Ï_" + str(Ï) + "_num_trials_" + str(num_trials) + "theta_T_3.tex")
plt.show()

# Size plots
plt.errorbar(Î´_list, np.mean(np.array(amp_size), axis = 1), yerr = np.std(np.array(amp_size), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(Î´_list, np.mean(np.array(theory_size), axis = 1),yerr = np.std(np.array(theory_size), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 5))
plt.xlabel("Î´")
plt.ylabel("Size")
plt.legend()
plt.savefig("figures/synthetic_estimation_linear_L4_" + "Ï_" + str(Ï) + "_num_trials_" + str(num_trials) + "theta_T_3_size.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_estimation_linear_L4_" + "Ï_" + str(Ï) + "_num_trials_" + str(num_trials) + "theta_T_3_size.tex")
plt.show()


```
