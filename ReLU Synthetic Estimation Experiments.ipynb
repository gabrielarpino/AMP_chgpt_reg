# ReLU size experiment (table) 1: κ = 100

```python
import numpy as np
import jax.numpy as jnp # Not used in any randomness, all randomness is in numpy
import numpy.random as nprandom
import matplotlib.pyplot as plt
from amp import norm_sq_corr, PAL
from amp.marginal_separable_jax_relu import q, GAMP_full
import amp.signal_configuration
from amp.signal_configuration import unif_prior_to_η_ϕ, unif_prior_to_η_ϕ_combinations
from amp.signal_priors import SignalPrior, SparseDiffSignal, SparseGaussianSignal, GaussianSignal
from amp.posterior_relu import η_to_ψ, posterior_over_η, MAP_η
from amp.signal_priors import GaussianSignal
from functools import partial
from tqdm import tqdm
tqdm = partial(tqdm, position=0, leave=True)

np.random.seed(700)

p = 600
σ = 0.3 # Not used by the functions of this script, but still passed in to maintain consistency with other code
L = 3 # num signals. L-1 changepoints

T = 10 # num_iterations
δ_list = [0.75, 1.25, 1.75] # TODO: run this experiment with [1.25, 1.75, 2.25]. Now with another reduced vmap, may be able to run more. Another idea is plot with a fixed δ, varying \kappa.
# δ_list = [0.25]
# δ_list = [1.0, 1.5]
num_trials = 10

# fractional_chgpt_locations = np.array([0.35, 0.7])
fractional_chgpt_locations = np.array([0.6])

corr_1_list = [[] for _ in range(len(δ_list))]
corr_2_list = [[] for _ in range(len(δ_list))]
corr_1_SE_list = [[] for _ in range(len(δ_list))]
corr_2_SE_list = [[] for _ in range(len(δ_list))]
fractional_η̂_list = [[] for _ in range(len(δ_list))]
amp_hausdorff = [[] for _ in range(len(δ_list))]
theory_hausdorff = [[] for _ in range(len(δ_list))]
amp_size = [[] for _ in range(len(δ_list))]
theory_size = [[] for _ in range(len(δ_list))]

for i in tqdm(range(len(δ_list))):
    δ = δ_list[i]
    print("δ = ", δ)

    n = int(δ * p)
    true_chgpt_locations = lambda n: (n*fractional_chgpt_locations).astype(int)

    # --- Preliminaries ---
    # Changepoint locations
    Δ = lambda n: int(n/5)
    Lmin = 1
    Lmax = L
    p_l = 1/L * np.ones(L) # puts equal weight on all number of signals, set None otherwise. Best for estimating size. 
    num_valid_configs, η_arr, p_η_arr, ϕ = unif_prior_to_η_ϕ_combinations(Lmin, Lmax, Δ(n), n, p_l = p_l) # Lxn matrix
    assert η_arr.shape == (num_valid_configs, Lmax-1) and \
        len(p_η_arr) == num_valid_configs
    assert ϕ.shape == (Lmax, n)

    print(f"True chgpts = {true_chgpt_locations(n)}")
    η_true = true_chgpt_locations(n)
    C_true = η_to_ψ(η_true, n)

    for trial_num in tqdm(range(num_trials)):

        # Gaussian Signal Prior
        B̃_cov = 1 * np.eye(L) * δ # Multiply by δ to keep the signal strength fixed
        B̃_cov[0, 1] = 0.1
        B̃_cov[1, 0] = 0.1
        signal_prior = GaussianSignal(B̃_cov)

        # Sparse Gaussian Signal Prior
        # α = 0.6 # fraction of nonzeros
        # σ_l_arr = np.array([0.5, 0.5]) # signal standard deviation
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)

        # Sparse Difference Signal Prior
        # α = 0.25 # fraction of nonzeros
        # # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # var_β_1 = 1
        # σ_w = 1
        # signal_prior = SparseDiffSignal(var_β_1, σ_w, α, L)

        # Soft threshold denoiser 
        # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # α = 0.5 # fraction of nonzeros
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)
        # st_ζ = 0.5 # The scaling factor in front of the effective noise std in the threshold
        # if soft thresholding. Setting st_ζ != None in the GAMP_full function.

        B̃ = signal_prior.sample(p)
        B̂_0 = signal_prior.sample(p)

        ρ = 1/δ * 1/p * B̃.T @ B̃

        # Generate the observation vector Y
        X = np.random.normal(0, np.sqrt(1/n), (n, p))
        Θ = X @ B̃
        ε = np.random.randn(n)*σ 
        Y = q(Θ, C_true, ε) 

        B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr, ν_fixed_arr, κ_T_fixed_arr, ν̂_fixed_arr, κ_B_fixed_arr = GAMP_full(B̃, δ, p, ϕ, σ, X, Y, C_true, T, B̂_0, signal_prior, signal_prior, st_ζ = None, verbose = False, seed = None, tqdm_disable = False)
        # B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr = GAMP(B̃, δ, p, ϕ, σ, X, Y, T, true_signal_prior=signal_prior, est_signal_prior=signal_prior)
        ν = ν_arr[-1]
        κ_T = κ_T_arr[-1]
        ν_fixed = ν_fixed_arr[-1]
        κ_T_fixed = κ_T_fixed_arr[-1]

        # (δ * ρ[1, 1] - δ * ν_arr[-1, 1, 1]) + (δ * ρ[0, 0] - δ * ν_arr[-1, 0, 0]) - 2*(δ * ρ[0, 1] - δ * ν_arr[-1, 0, 1])
        # ν_fixed, κ_T_fixed = SE_fixed_C(C_true, signal_prior, signal_prior, δ, p, ϕ, L, σ, T, tqdm_disable = False, st_ζ=None)

        corr_1_list[i].append(norm_sq_corr(B̂[:, 0], B̃[:, 0]))
        corr_2_list[i].append(norm_sq_corr(B̂[:, 1], B̃[:, 1]))
        corr_1_SE_list[i].append(ν_fixed[0, 0]**2 / ( ρ[0, 0] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[0, 0]))
        corr_2_SE_list[i].append(ν_fixed[1, 1]**2 / ( ρ[1, 1] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[1, 1]))

        ### --- Approx posterior --- ###
        # Compute approximate posterior MAP
        print("Beginning approximate posterior computation...")
        ψ_arr_amp, η_idx_amp = amp.signal_configuration.η_to_ψ_jax_combinations_mapped(η_arr, n, L)       
        post_amp = amp.posterior_relu.compute_approx_posterior(ψ_arr_amp, Θ_t, Y, n, ρ, σ, ν, κ_T)
        map_idx_amp = np.argmax(post_amp)
        ψ̂ = ψ_arr_amp[map_idx_amp]
        η̂ = η_arr[η_idx_amp[map_idx_amp]]
        η̂ = η̂[η̂ > 0]

        fractional_η̂_list[i].append(η̂ / n)
        amp_size[i].append(len(η̂))
        amp_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(η̂, η_true), 0.5)) # Sometimes infs occur because no change point is returned

        # --- Compute η̂ from theoretical ---
        Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), ρ, size=n))).reshape(n, L)
        # Sample V̄_Θ
        V_Θ = (Z @ np.linalg.inv(ρ) @ ν_fixed + jnp.array(nprandom.multivariate_normal(
            jnp.zeros(L), κ_T_fixed, size=n))).reshape(n, L)
        # post_V_q = posterior_over_η(η_arr, p_η_arr, V_Θ_i, q(Z_i, C_true, σ), jnp.array(ρ), σ, ν, κ_T)

        ψ_arr_theory, η_idx_theory = amp.signal_configuration.η_to_ψ_jax_combinations_mapped(η_arr, n, L)       
        post_theory = amp.posterior_relu.compute_approx_posterior(ψ_arr_theory, V_Θ, q(Z, C_true, ε), n, ρ, σ, ν, κ_T)
        map_idx_theory = np.argmax(post_theory)
        ψ̂_theory = ψ_arr_theory[map_idx_theory]
        η̂_theory = η_arr[η_idx_theory[map_idx_theory]]
        η̂_theory = η̂_theory[η̂_theory > 0] # Remove negative ones, which are invalid

        theory_size[i].append(len(η̂_theory))
        theory_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(η̂_theory, η_true), 0.5)) # Sometimes infs occur because no change point is returned
        print("Approximate posterior computation finished.")

        ### --- Plotting --- ###
        fig = plt.figure()
        plt.xlim((0, n))
        plt.plot(np.arange(n), ψ̂, label="AMP")
        plt.plot(np.arange(n), ψ̂_theory, label="Theory", ls="--")
        [plt.axvline(x=loc, color='r', linestyle='-.', alpha = 0.5, label="AMP chgpt") for loc in η̂]
        [plt.axvline(x=loc, color='y', linestyle='-.', alpha = 0.5, label="Theory chgpt") for loc in η̂_theory]
        [plt.axvline(x=c, color='black', linestyle='--', alpha = 0.3, label="True") for c in true_chgpt_locations(n)]
        plt.legend()
        plt.show()

# Fix for tikzplotlib
from matplotlib.lines import Line2D
from matplotlib.legend import Legend
Line2D._us_dashSeq    = property(lambda self: self._dash_pattern[1])
Line2D._us_dashOffset = property(lambda self: self._dash_pattern[0])
Legend._ncol = property(lambda self: self._ncols)

# Sqcorr plots
plt.errorbar(δ_list, np.mean(np.array(corr_1_list), axis = 1), yerr = np.std(np.array(corr_1_list), axis = 1) , label="AMP β¹", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np.array(corr_2_list), axis = 1),yerr = np.std(np.array(corr_2_list), axis = 1) ,  label="AMP β²", marker="o", ls="--", c=PAL[2], alpha = 0.75)
plt.errorbar(δ_list,np.mean(np.array(corr_1_SE_list), axis = 1),yerr = np.std(np.array(corr_1_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[0])
plt.errorbar(δ_list, np.mean(np.array(corr_2_SE_list), axis = 1),yerr = np.std(np.array(corr_2_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[2])
plt.ylim((0, 1))
plt.xlabel("δ")
plt.ylabel("Norm Sq Corr")
plt.legend()
plt.show()

# Hausdorff plots
np_amp_hausdorff = np.array(amp_hausdorff)
np_theory_hausdorff = np.array(theory_hausdorff)
plt.errorbar(δ_list, np.mean(np_amp_hausdorff, axis = 1), yerr = np.mean(np.array(amp_hausdorff), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np_theory_hausdorff, axis = 1),yerr = np.mean(np.array(theory_hausdorff), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 0.5))
plt.xlabel("δ")
plt.ylabel("Hausdorff distance")
plt.legend()
plt.savefig("figures/synthetic_estimation_linear_L3_" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + "theta_T_3.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_estimation_linear_L3_" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + "theta_T_3.tex")
plt.show()

# Size plots
plt.errorbar(δ_list, np.mean(np.array(amp_size), axis = 1), yerr = np.std(np.array(amp_size), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np.array(theory_size), axis = 1),yerr = np.std(np.array(theory_size), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 5))
plt.xlabel("δ")
plt.ylabel("Size")
plt.legend()
plt.savefig("figures/synthetic_size_relu_L3_redo" + "_num_trials_" + str(num_trials) + "theta_T_3_size.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_size_relu_L3_redo" + "_num_trials_" + str(num_trials) + "theta_T_3_size.tex")
plt.show()

print("AMP size standard deviation: ", np.std(np.array(amp_size), axis = 1))
```

```python
import numpy as np
import jax.numpy as jnp # Not used in any randomness, all randomness is in numpy
import numpy.random as nprandom
import matplotlib.pyplot as plt
from amp import norm_sq_corr, PAL
from amp.marginal_separable_jax_relu import q, GAMP_full
import amp.signal_configuration
from amp.signal_configuration import unif_prior_to_η_ϕ, unif_prior_to_η_ϕ_combinations
from amp.signal_priors import SignalPrior, SparseDiffSignal, SparseGaussianSignal, GaussianSignal
from amp.posterior_relu import η_to_ψ, posterior_over_η, MAP_η
from amp.signal_priors import GaussianSignal
from functools import partial
from tqdm import tqdm
tqdm = partial(tqdm, position=0, leave=True)

np.random.seed(800)

p = 600
σ = 0.5 # Not used by the functions of this script, but still passed in to maintain consistency with other code
L = 3 # num signals. L-1 changepoints

T = 10 # num_iterations
δ_list = [0.75, 1.25, 1.75] # TODO: run this experiment with [1.25, 1.75, 2.25]. Now with another reduced vmap, may be able to run more. Another idea is plot with a fixed δ, varying \kappa.
# δ_list = [0.25]
# δ_list = [1.0, 1.5]
num_trials = 10

# fractional_chgpt_locations = np.array([0.35, 0.7])
fractional_chgpt_locations = np.array([0.6])

corr_1_list = [[] for _ in range(len(δ_list))]
corr_2_list = [[] for _ in range(len(δ_list))]
corr_1_SE_list = [[] for _ in range(len(δ_list))]
corr_2_SE_list = [[] for _ in range(len(δ_list))]
fractional_η̂_list = [[] for _ in range(len(δ_list))]
amp_hausdorff = [[] for _ in range(len(δ_list))]
theory_hausdorff = [[] for _ in range(len(δ_list))]
amp_size = [[] for _ in range(len(δ_list))]
theory_size = [[] for _ in range(len(δ_list))]

for i in tqdm(range(len(δ_list))):
    δ = δ_list[i]
    print("δ = ", δ)

    n = int(δ * p)
    true_chgpt_locations = lambda n: (n*fractional_chgpt_locations).astype(int)

    # --- Preliminaries ---
    # Changepoint locations
    Δ = lambda n: int(n/5)
    Lmin = 1
    Lmax = L
    p_l = 1/L * np.ones(L) # puts equal weight on all number of signals, set None otherwise. Best for estimating size. 
    num_valid_configs, η_arr, p_η_arr, ϕ = unif_prior_to_η_ϕ_combinations(Lmin, Lmax, Δ(n), n, p_l = p_l) # Lxn matrix
    assert η_arr.shape == (num_valid_configs, Lmax-1) and \
        len(p_η_arr) == num_valid_configs
    assert ϕ.shape == (Lmax, n)

    print(f"True chgpts = {true_chgpt_locations(n)}")
    η_true = true_chgpt_locations(n)
    C_true = η_to_ψ(η_true, n)

    for trial_num in tqdm(range(num_trials)):

        # Gaussian Signal Prior
        B̃_cov = 1 * np.eye(L) * δ
        B̃_cov[0, 1] = 0.1
        B̃_cov[1, 0] = 0.1
        signal_prior = GaussianSignal(B̃_cov)

        # Sparse Gaussian Signal Prior
        # α = 0.6 # fraction of nonzeros
        # σ_l_arr = np.array([0.5, 0.5]) # signal standard deviation
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)

        # Sparse Difference Signal Prior
        # α = 0.25 # fraction of nonzeros
        # # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # var_β_1 = 1
        # σ_w = 1
        # signal_prior = SparseDiffSignal(var_β_1, σ_w, α, L)

        # Soft threshold denoiser 
        # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # α = 0.5 # fraction of nonzeros
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)
        # st_ζ = 0.5 # The scaling factor in front of the effective noise std in the threshold
        # if soft thresholding. Setting st_ζ != None in the GAMP_full function.

        B̃ = signal_prior.sample(p)
        B̂_0 = signal_prior.sample(p)

        ρ = 1/δ * 1/p * B̃.T @ B̃

        # Generate the observation vector Y
        X = np.random.normal(0, np.sqrt(1/n), (n, p))
        Θ = X @ B̃
        ε = np.random.rand(n) # uniform dist
        Y = q(Θ, C_true, ε) 

        B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr, ν_fixed_arr, κ_T_fixed_arr, ν̂_fixed_arr, κ_B_fixed_arr = GAMP_full(B̃, δ, p, ϕ, σ, X, Y, C_true, T, B̂_0, signal_prior, signal_prior, st_ζ = None, verbose = False, seed = None, tqdm_disable = False)
        # B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr = GAMP(B̃, δ, p, ϕ, σ, X, Y, T, true_signal_prior=signal_prior, est_signal_prior=signal_prior)
        ν = ν_arr[-1]
        κ_T = κ_T_arr[-1]
        ν_fixed = ν_fixed_arr[-1]
        κ_T_fixed = κ_T_fixed_arr[-1]

        # (δ * ρ[1, 1] - δ * ν_arr[-1, 1, 1]) + (δ * ρ[0, 0] - δ * ν_arr[-1, 0, 0]) - 2*(δ * ρ[0, 1] - δ * ν_arr[-1, 0, 1])
        # ν_fixed, κ_T_fixed = SE_fixed_C(C_true, signal_prior, signal_prior, δ, p, ϕ, L, σ, T, tqdm_disable = False, st_ζ=None)

        corr_1_list[i].append(norm_sq_corr(B̂[:, 0], B̃[:, 0]))
        corr_2_list[i].append(norm_sq_corr(B̂[:, 1], B̃[:, 1]))
        corr_1_SE_list[i].append(ν_fixed[0, 0]**2 / ( ρ[0, 0] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[0, 0]))
        corr_2_SE_list[i].append(ν_fixed[1, 1]**2 / ( ρ[1, 1] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[1, 1]))

        ### --- Approx posterior --- ###
        # Compute approximate posterior MAP
        print("Beginning approximate posterior computation...")
        ψ_arr_amp, η_idx_amp = amp.signal_configuration.η_to_ψ_jax_combinations_mapped(η_arr, n, L)       
        post_amp = amp.posterior_relu.compute_approx_posterior(ψ_arr_amp, Θ_t, Y, n, ρ, σ, ν, κ_T)
        map_idx_amp = np.argmax(post_amp)
        ψ̂ = ψ_arr_amp[map_idx_amp]
        η̂ = η_arr[η_idx_amp[map_idx_amp]]
        η̂ = η̂[η̂ > 0]

        fractional_η̂_list[i].append(η̂ / n)
        amp_size[i].append(len(η̂))
        amp_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(η̂, η_true), 0.5)) # Sometimes infs occur because no change point is returned

        # --- Compute η̂ from theoretical ---
        Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), ρ, size=n))).reshape(n, L)
        # Sample V̄_Θ
        V_Θ = (Z @ np.linalg.inv(ρ) @ ν_fixed + jnp.array(nprandom.multivariate_normal(
            jnp.zeros(L), κ_T_fixed, size=n))).reshape(n, L)
        # post_V_q = posterior_over_η(η_arr, p_η_arr, V_Θ_i, q(Z_i, C_true, σ), jnp.array(ρ), σ, ν, κ_T)

        ψ_arr_theory, η_idx_theory = amp.signal_configuration.η_to_ψ_jax_combinations_mapped(η_arr, n, L)       
        post_theory = amp.posterior_relu.compute_approx_posterior(ψ_arr_theory, V_Θ, q(Z, C_true, ε), n, ρ, σ, ν, κ_T)
        map_idx_theory = np.argmax(post_theory)
        ψ̂_theory = ψ_arr_theory[map_idx_theory]
        η̂_theory = η_arr[η_idx_theory[map_idx_theory]]
        η̂_theory = η̂_theory[η̂_theory > 0] # Remove negative ones, which are invalid

        theory_size[i].append(len(η̂_theory))
        theory_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(η̂_theory, η_true), 0.5)) # Sometimes infs occur because no change point is returned
        print("Approximate posterior computation finished.")

        ### --- Plotting --- ###
        fig = plt.figure()
        plt.xlim((0, n))
        plt.plot(np.arange(n), ψ̂, label="AMP")
        plt.plot(np.arange(n), ψ̂_theory, label="Theory", ls="--")
        [plt.axvline(x=loc, color='r', linestyle='-.', alpha = 0.5, label="AMP chgpt") for loc in η̂]
        [plt.axvline(x=loc, color='y', linestyle='-.', alpha = 0.5, label="Theory chgpt") for loc in η̂_theory]
        [plt.axvline(x=c, color='black', linestyle='--', alpha = 0.3, label="True") for c in true_chgpt_locations(n)]
        plt.legend()
        plt.show()

# Fix for tikzplotlib
from matplotlib.lines import Line2D
from matplotlib.legend import Legend
Line2D._us_dashSeq    = property(lambda self: self._dash_pattern[1])
Line2D._us_dashOffset = property(lambda self: self._dash_pattern[0])
Legend._ncol = property(lambda self: self._ncols)

# Sqcorr plots
plt.errorbar(δ_list, np.mean(np.array(corr_1_list), axis = 1), yerr = np.std(np.array(corr_1_list), axis = 1) , label="AMP β¹", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np.array(corr_2_list), axis = 1),yerr = np.std(np.array(corr_2_list), axis = 1) ,  label="AMP β²", marker="o", ls="--", c=PAL[2], alpha = 0.75)
plt.errorbar(δ_list,np.mean(np.array(corr_1_SE_list), axis = 1),yerr = np.std(np.array(corr_1_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[0])
plt.errorbar(δ_list, np.mean(np.array(corr_2_SE_list), axis = 1),yerr = np.std(np.array(corr_2_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[2])
plt.ylim((0, 1))
plt.xlabel("δ")
plt.ylabel("Norm Sq Corr")
plt.legend()
plt.show()

# Hausdorff plots
np_amp_hausdorff = np.array(amp_hausdorff)
np_theory_hausdorff = np.array(theory_hausdorff)
plt.errorbar(δ_list, np.mean(np_amp_hausdorff, axis = 1), yerr = np.mean(np.array(amp_hausdorff), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np_theory_hausdorff, axis = 1),yerr = np.mean(np.array(theory_hausdorff), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 0.5))
plt.xlabel("δ")
plt.ylabel("Hausdorff distance")
plt.legend()
plt.savefig("figures/synthetic_estimation_linear_L3_table2" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + "theta_T_3.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_estimation_linear_L3_table2" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + "theta_T_3.tex")
plt.show()

# Size plots
plt.errorbar(δ_list, np.mean(np.array(amp_size), axis = 1), yerr = np.std(np.array(amp_size), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np.array(theory_size), axis = 1),yerr = np.std(np.array(theory_size), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 5))
plt.xlabel("δ")
plt.ylabel("Size")
plt.legend()
plt.savefig("figures/synthetic_size_relu_L3_table2" + "_num_trials_" + str(num_trials) + "theta_T_3_size.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_size_relu_L3_table2" + "_num_trials_" + str(num_trials) + "theta_T_3_size.tex")
plt.show()

print("AMP size standard deviation: ", np.std(np.array(amp_size), axis = 1))
```

```python
import numpy as np
import jax.numpy as jnp # Not used in any randomness, all randomness is in numpy
import numpy.random as nprandom
import matplotlib.pyplot as plt
from amp import norm_sq_corr, PAL
from amp.marginal_separable_jax_relu import q, GAMP_full
import amp.signal_configuration
from amp.signal_configuration import unif_prior_to_η_ϕ, unif_prior_to_η_ϕ_combinations
from amp.signal_priors import SignalPrior, SparseDiffSignal, SparseGaussianSignal, GaussianSignal
from amp.posterior_relu import η_to_ψ, posterior_over_η, MAP_η
from amp.signal_priors import GaussianSignal
from functools import partial
from tqdm import tqdm
tqdm = partial(tqdm, position=0, leave=True)

np.random.seed(657)

p = 600
σ = 0.1 # Not used by the functions of this script, but still passed in to maintain consistency with other code
L = 3 # num signals. L-1 changepoints

T = 10 # num_iterations
δ_list = [0.75, 1.25, 1.75, 2.25]
# δ_list = [1.0, 1.5]
num_trials = 10

fractional_chgpt_locations = np.array([0.35, 0.7])
# fractional_chgpt_locations = np.array([0.6])

corr_1_list = [[] for _ in range(len(δ_list))]
corr_2_list = [[] for _ in range(len(δ_list))]
corr_1_SE_list = [[] for _ in range(len(δ_list))]
corr_2_SE_list = [[] for _ in range(len(δ_list))]
fractional_η̂_list = [[] for _ in range(len(δ_list))]
amp_hausdorff = [[] for _ in range(len(δ_list))]
theory_hausdorff = [[] for _ in range(len(δ_list))]
amp_size = [[] for _ in range(len(δ_list))]
theory_size = [[] for _ in range(len(δ_list))]

for i in tqdm(range(len(δ_list))):
    δ = δ_list[i]
    print("δ = ", δ)

    n = int(δ * p)
    true_chgpt_locations = lambda n: (n*fractional_chgpt_locations).astype(int)

    # --- Preliminaries ---
    # Changepoint locations
    Δ = lambda n: int(n/5)
    Lmin = L
    Lmax = L
    # p_l = 1/L * np.ones(L) # puts equal weight on all number of signals, set None otherwise. Best for estimating size. 
    p_l = None 
    num_valid_configs, η_arr, p_η_arr, ϕ = unif_prior_to_η_ϕ(Lmin, Lmax, Δ(n), n, p_l = p_l) # Lxn matrix
    assert η_arr.shape == (num_valid_configs, Lmax-1) and \
        len(p_η_arr) == num_valid_configs
    assert ϕ.shape == (Lmax, n)

    print(f"True chgpts = {true_chgpt_locations(n)}")
    η_true = true_chgpt_locations(n)
    C_true = η_to_ψ(η_true, n)

    for trial_num in tqdm(range(num_trials)):

        # Gaussian Signal Prior
        B̃_cov = 15 * np.eye(L)
        B̃_cov[0, 1] = 0.2
        B̃_cov[1, 0] = 0.2
        signal_prior = GaussianSignal(B̃_cov)

        # Sparse Gaussian Signal Prior
        # α = 0.6 # fraction of nonzeros
        # σ_l_arr = np.array([0.5, 0.5]) # signal standard deviation
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)

        # Sparse Difference Signal Prior
        # α = 0.25 # fraction of nonzeros
        # # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # var_β_1 = 1
        # σ_w = 1
        # signal_prior = SparseDiffSignal(var_β_1, σ_w, α, L)

        # Soft threshold denoiser 
        # σ_l_arr = np.array([2.5, 2.5]) # signal standard deviation
        # α = 0.5 # fraction of nonzeros
        # signal_prior = SparseGaussianSignal(α, δ, σ_l_arr)
        # st_ζ = 0.5 # The scaling factor in front of the effective noise std in the threshold
        # if soft thresholding. Setting st_ζ != None in the GAMP_full function.

        B̃ = signal_prior.sample(p)
        B̂_0 = signal_prior.sample(p)

        ρ = 1/δ * 1/p * B̃.T @ B̃

        # Generate the observation vector Y
        X = np.random.normal(0, np.sqrt(1/n), (n, p))
        Θ = X @ B̃
        ε = np.random.rand(n) # uniform dist
        Y = q(Θ, C_true, ε) 

        B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr, ν_fixed_arr, κ_T_fixed_arr, ν̂_fixed_arr, κ_B_fixed_arr = GAMP_full(B̃, δ, p, ϕ, σ, X, Y, C_true, T, B̂_0, signal_prior, signal_prior, st_ζ = None, verbose = False, seed = None, tqdm_disable = False)
        # B̂, Θ_t, ν_arr, κ_T_arr, ν̂_arr, κ_B_arr = GAMP(B̃, δ, p, ϕ, σ, X, Y, T, true_signal_prior=signal_prior, est_signal_prior=signal_prior)
        ν = ν_arr[-1]
        κ_T = κ_T_arr[-1]
        ν_fixed = ν_fixed_arr[-1]
        κ_T_fixed = κ_T_fixed_arr[-1]

        # (δ * ρ[1, 1] - δ * ν_arr[-1, 1, 1]) + (δ * ρ[0, 0] - δ * ν_arr[-1, 0, 0]) - 2*(δ * ρ[0, 1] - δ * ν_arr[-1, 0, 1])
        # ν_fixed, κ_T_fixed = SE_fixed_C(C_true, signal_prior, signal_prior, δ, p, ϕ, L, σ, T, tqdm_disable = False, st_ζ=None)

        corr_1_list[i].append(norm_sq_corr(B̂[:, 0], B̃[:, 0]))
        corr_2_list[i].append(norm_sq_corr(B̂[:, 1], B̃[:, 1]))
        corr_1_SE_list[i].append(ν_fixed[0, 0]**2 / ( ρ[0, 0] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[0, 0]))
        corr_2_SE_list[i].append(ν_fixed[1, 1]**2 / ( ρ[1, 1] * (κ_T_fixed + ν_fixed.T @ np.linalg.inv(ρ) @ ν_fixed)[1, 1]))

        ### --- Approx posterior --- ###
        # Compute approximate posterior MAP
        print("Beginning approximate posterior computation...")
        ψ_arr_amp, η_idx_amp = amp.signal_configuration.η_to_ψ_jax_combinations_mapped(η_arr, n, L)       
        post_amp = amp.posterior_relu.compute_approx_posterior(ψ_arr_amp, Θ_t, Y, n, ρ, σ, ν, κ_T)
        map_idx_amp = np.argmax(post_amp)
        ψ̂ = ψ_arr_amp[map_idx_amp]
        η̂ = η_arr[η_idx_amp[map_idx_amp]]
        η̂ = η̂[η̂ > 0]

        fractional_η̂_list[i].append(η̂ / n)
        amp_size[i].append(len(η̂))
        amp_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(η̂, η_true), 0.5)) # Sometimes infs occur because no change point is returned

        # --- Compute η̂ from theoretical ---
        Z = (np.array(nprandom.multivariate_normal(jnp.zeros(L), ρ, size=n))).reshape(n, L)
        # Sample V̄_Θ
        V_Θ = (Z @ np.linalg.inv(ρ) @ ν_fixed + jnp.array(nprandom.multivariate_normal(
            jnp.zeros(L), κ_T_fixed, size=n))).reshape(n, L)
        # post_V_q = posterior_over_η(η_arr, p_η_arr, V_Θ_i, q(Z_i, C_true, σ), jnp.array(ρ), σ, ν, κ_T)

        ψ_arr_theory, η_idx_theory = amp.signal_configuration.η_to_ψ_jax_combinations_mapped(η_arr, n, L)       
        post_theory = amp.posterior_relu.compute_approx_posterior(ψ_arr_theory, V_Θ, q(Z, C_true, ε), n, ρ, σ, ν, κ_T)
        map_idx_theory = np.argmax(post_theory)
        ψ̂_theory = ψ_arr_theory[map_idx_theory]
        η̂_theory = η_arr[η_idx_theory[map_idx_theory]]
        η̂_theory = η̂_theory[η̂_theory > 0] # Remove negative ones, which are invalid

        theory_size[i].append(len(η̂_theory))
        theory_hausdorff[i].append(min((1/n) * amp.performance_measures.hausdorff(η̂_theory, η_true), 0.5)) # Sometimes infs occur because no change point is returned
        print("Approximate posterior computation finished.")

        ### --- Plotting --- ###
        fig = plt.figure()
        plt.xlim((0, n))
        plt.plot(np.arange(n), ψ̂, label="AMP")
        plt.plot(np.arange(n), ψ̂_theory, label="Theory", ls="--")
        [plt.axvline(x=loc, color='r', linestyle='-.', alpha = 0.5, label="AMP chgpt") for loc in η̂]
        [plt.axvline(x=loc, color='y', linestyle='-.', alpha = 0.5, label="Theory chgpt") for loc in η̂_theory]
        [plt.axvline(x=c, color='black', linestyle='--', alpha = 0.3, label="True") for c in true_chgpt_locations(n)]
        plt.legend()
        plt.show()

# Fix for tikzplotlib
from matplotlib.lines import Line2D
from matplotlib.legend import Legend
Line2D._us_dashSeq    = property(lambda self: self._dash_pattern[1])
Line2D._us_dashOffset = property(lambda self: self._dash_pattern[0])
Legend._ncol = property(lambda self: self._ncols)

# Sqcorr plots
plt.errorbar(δ_list, np.mean(np.array(corr_1_list), axis = 1), yerr = np.std(np.array(corr_1_list), axis = 1) , label="AMP β¹", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np.array(corr_2_list), axis = 1),yerr = np.std(np.array(corr_2_list), axis = 1) ,  label="AMP β²", marker="o", ls="--", c=PAL[2], alpha = 0.75)
plt.errorbar(δ_list,np.mean(np.array(corr_1_SE_list), axis = 1),yerr = np.std(np.array(corr_1_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[0])
plt.errorbar(δ_list, np.mean(np.array(corr_2_SE_list), axis = 1),yerr = np.std(np.array(corr_2_SE_list), axis = 1) ,  label="Theory", marker="x", alpha = 0.75, c=PAL[2])
plt.ylim((0, 1))
plt.xlabel("δ")
plt.ylabel("Norm Sq Corr")
plt.legend()
plt.show()

# Hausdorff plots
np_amp_hausdorff = np.array(amp_hausdorff)
np_theory_hausdorff = np.array(theory_hausdorff)
plt.errorbar(δ_list, np.mean(np_amp_hausdorff, axis = 1), yerr = np.mean(np.array(amp_hausdorff), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
plt.errorbar(δ_list, np.mean(np_theory_hausdorff, axis = 1),yerr = np.mean(np.array(theory_hausdorff), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
plt.ylim((0, 0.5))
plt.xlabel("δ")
plt.ylabel("Hausdorff distance")
plt.legend()
plt.savefig("figures/synthetic_estimation_relu_L3_2" + "_num_trials_" + str(num_trials) + "theta_T_3.pdf")
import tikzplotlib
tikzplotlib.save("figures/synthetic_estimation_relu_L3_2" + "_num_trials_" + str(num_trials) + "theta_T_3.tex")
plt.show()

# Size plots
# plt.errorbar(δ_list, np.mean(np.array(amp_size), axis = 1), yerr = np.std(np.array(amp_size), axis = 1), label="AMP", marker="o", ls="--", c=PAL[0], alpha = 0.75)
# plt.errorbar(δ_list, np.mean(np.array(theory_size), axis = 1),yerr = np.std(np.array(theory_size), axis = 1),  label="Theory", marker="x", alpha = 3/4, c=PAL[1])
# plt.ylim((0, 5))
# plt.xlabel("δ")
# plt.ylabel("Size")
# plt.legend()
# plt.savefig("figures/synthetic_estimation_linear_L4_" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + "theta_T_3_size.pdf")
# import tikzplotlib
# tikzplotlib.save("figures/synthetic_estimation_linear_L4_" + "σ_" + str(σ) + "_num_trials_" + str(num_trials) + "theta_T_3_size.tex")
# plt.show()
```
